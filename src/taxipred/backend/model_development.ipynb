```python
import pandas as pd

df = pd.read_csv("../data/cleaned_train_data.csv")

df.head(10)

categorical_cols = ['Time_of_Day', 'Day_of_Week', 'Traffic_Conditions', 'Weather']

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df.head(20)
```

```python
df_encoded
```

```python
df.describe().T
```

```python
from sklearn.model_selection import train_test_split

x = df_encoded.drop('Trip_Price', axis=1) # Använd den nya DataFrame
y = df_encoded['Trip_Price']
x.head(15)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_test, y_pred)
```

Beräknar MAE. MAE mäter den genomsnittliga storleken på felet i dina förutsägelser, utan att bry sig om riktningen (om gissningen var för hög eller för låg).

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

Beräknar RMSE = Root Mean Squared Error (RMSE). RMSE ger större vikt åt stora fel (stora outliers) än vad MAE gör, vilket är bra om du vill straffa modellen hårt för väldigt dåliga gissningar.

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

## Linjär Regression model 

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_y_pred = lr_model.predict(X_test)

lr_mae = mean_absolute_error(y_test, lr_y_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_y_pred))



print("Resultat: Linjär Regression")
print(f"Mean Absolute Error (MAE): {lr_mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {lr_rmse:.2f}")
```

## GradientBoostingRegressor model

```python
from sklearn.ensemble import GradientBoostingRegressor


gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr_model.fit(X_train, y_train)

gbr_y_pred = gbr_model.predict(X_test)

gbr_mae = mean_absolute_error(y_test, gbr_y_pred)
gbr_rmse = np.sqrt(mean_squared_error(y_test, gbr_y_pred))

print("Resultat: Gradient Boosting Regression")
print(f"Mean Absolute Error (MAE): {gbr_mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {gbr_rmse:.2f}")
```

## GradientBoostingRegressor är vinnaren av modellerna 

- Träna nya modellen med all data
- Exportera modellen med joblib
 

```python
import joblib

# parametrarna hjälper modellen få bättre resultat och ger den bättre prestanda
gbr_model_final = GradientBoostingRegressor(
    n_estimators=100,  # Antal träd att bygga (standard är 100).
    learning_rate=0.1, # Hur snabbt modellen lär sig från sina fel (standard är 0.1).
    max_depth=3, # Hur djupt varje individuellt träd får bli (begränsar komplexiteten och förhindrar överanpassning, standard är 3).
    random_state=42 # Ger dig ett reproducerbart resultat varje gång du kör koden.
    )

gbr_model_final.fit(x, y) 

MODEL_PATH = "../models/gbr_final_model.joblib" 
joblib.dump(gbr_model_final, MODEL_PATH)
```
