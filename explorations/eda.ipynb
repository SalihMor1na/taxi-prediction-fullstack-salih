```python
import pandas as pd
from taxipred.utils.constants import TAXI_CSV_PATH

df = pd.read_csv(str(TAXI_CSV_PATH))

df.head(10)
```

```python
print(df.isnull().sum())
```

```python
future_user_input = df[df['Trip_Price'].isnull()].copy()
len(future_user_input)
```

```python
df.dropna(inplace=True) 
len(df)
```

```python
df.head(15)
```

```python
## Få ut outliers med qutanile 0.97 = 97% av högsta värdet
outliers = df['Trip_Price'].quantile(0.97)
```

```python
df = df[df['Trip_Price'] < outliers].copy()
len(df)
```

## Exporterar den nya rena datan till data folder

```python
from pathlib import Path


raw_data_path = TAXI_CSV_PATH

## Använder path för att hitta i vilken folder taxi_csv_path är lagrad
DATA_DIR = Path(str(raw_data_path)).parent 

EXPORT_PATH = DATA_DIR / "cleaned_train_data.csv"

df.to_csv(EXPORT_PATH, index=False) 
```
